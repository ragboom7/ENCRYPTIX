```python
print('welcome to our chatbot service. Once your queries are cleared, type:exit, goodbye, bye or quit to go out of the chatbot service. Thanks!')
import re

class Chatbot:
    def __init__(self):
        self.rules = [
            (re.compile(r'(hello|hi|hey|wassup)', re.IGNORECASE), self.greet),
            (re.compile(r'(store\s+open|opening|operating|store\s+open\s+\w+\s+store)', re.IGNORECASE), self.open),
            (re.compile(r'(store\s+locations|locations\s+\w+\s+store|stores\s+\w+\s+located|store\s+\w+\s+branches|branches\s+\w+\s+store|locations|location|located)', re.IGNORECASE), self.locations),
            (re.compile(r'(return\s+\w+\s+item|return|item\s+\w+\s+return)', re.IGNORECASE), self.returnp),
            (re.compile(r'\b(delivery\s+(date|information|details)\s+\w+)\b|delivery|\b(date\s+of\s+delivery)\b|\b(track\s+(order|package|shipment)\s+\w+)\b|\b(order\s+(status|delivery|info)\s+\w+)\b|\b(shipping\s+(info|information|details)\s+\w+)\b|\b(date\s+delivery)\b', re.IGNORECASE), self.track),
            (re.compile(r'(thanks|thank\s+you)', re.IGNORECASE), self.thanks)
        ]
        self.default_response = "I am sorry, I do not understand. Could you please rephrase that or ask a different question?"

    def greet(self):
        return 'Hello! How can I help you?'

    def open(self):
        return 'Our store is open from 9am to 8pm Monday to Saturday. Sundays are off.'

    def locations(self):
        return 'Our store has locations in Chennai, Bangalore, Kolkata, Delhi, and Hyderabad.'

    def track(self):
        return ('If you wish to track your order and find the delivery date, go to the main menu and click on the Track order button. '
                'Enter the billing code and press enter. This will help you track your package!')

    def returnp(self):
        return ('If you wish to return your order, go to the main menu and click on the return order button. '
                'Enter the billing code and press enter. This will allow you to return your order.')

    def thanks(self):
        return 'You are welcome! Come back soon!'

    def run(self, user_input):
        #user_input = user_input.lower()
        for pattern, response in self.rules:
            match = pattern.search(user_input)
            if match:
                return response()
        return self.default_response

    def user(self):
        print('Chatbot: Hello! How can I help you?')
        while True:
            user_input = input('You: ')
            if user_input.lower() in ["exit", "quit", "bye", "goodbye"]:
                print("Chatbot: Goodbye! Thank you for visiting our online store.")
                break
            value = self.run(user_input)
            print('Chatbot:', value)

store = Chatbot()
store.user()
```
